# Scraping Tokopedia Reviews
(Which apparently has a good sentiment analysis!) <br/>
Learning source: https://www.dataquest.io/blog/web-scraping-beautifulsoup/

```python
from bs4 import BeautifulSoup
from requests import get
```

```python
# Extracting the webpage
url = 'https://www.tokopedia.com/ray189/oppo-a37?trkid=f=Ca0000L000P0W0S0Sh00Co0Po0Fr0Cb0_src=search_page=1_ob=23_q=oppo_po=2_catid=24'

response = get(url)
# print(response.text)
```

```python
html_soup = BeautifulSoup(response.text, 'html.parser')

html_soup
```

```python
review_containers = html_soup.find_all('ul', class_='review-container')
review_containers
```
